2. Пусть задана плоскость: A1x + B1y + C1z + D1 = 0
и прямая:
```
x-x1/x2-x1 = y-y1/y2-y1 = z-z1/z2-z1
Как узнать, принадлежит прямая плоскости или нет?
```

Из уравнения прямой мы знаем координаты минимум двух точек A1(x1, y1, z1) и A2(x2,y2,z2), принадлежащих этой прямой.

***Решите систему уравнений и неравенств:***
```
exp(x) + x∙(1 – y) = 1  
y = x2 – 1
------------------
y = x2 – 1
exp(x) + x – xy = 1 -> y = (exp(x) + x –  1) / x
```

````python
x = np.linspace(-2, 3, 201)

plt.figure(figsize = (6, 12))
plt.plot(x, (np.exp(x)+ x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.axis('scaled')
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
````
