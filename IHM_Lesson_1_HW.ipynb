```python
from sympy import *
```

```python
x, y, z = symbols('x y z')
```

```python
A = Matrix([[3, -2, 5], [7, 4, 8], [5, -3, -4]])
b = Matrix([7, 3, -12])
```

```python
linsolve((A, b), [x, y, z])
```
